```python
import json

from openai import OpenAI
from colorama import init, Fore
from loguru import logger

init(autoreset=True)
client = OpenAI(
  base_url="http://localhost:11434/v1/",
  api_key = "not need key"
)

def get_completion(query: str, stream=False, max_retry=5):
    params = dict(model="llama3:70b", messages=[{"role": "user", "content": query}], stream=stream, max_tokens=8192, temperature=0.8, top_p=0.8)

    response = client.chat.completions.create(**params)

    for _ in range(max_retry):
        if not stream:
            reply = response.choices[0].message.content
            logger.info(f"Final Reply: \n{reply}")
            # print(reply)
            return

        else:
            output = ""
            for chunk in response:
                content = chunk.choices[0].delta.content or ""
                print(Fore.BLUE + content, end="", flush=True)
                output += content

                if chunk.choices[0].finish_reason == "stop":
                    return

        response = client.chat.completions.create(**params)
```

```python
query = "现在你是一名优秀的南京地导，接下来的一个任务是为大家介绍一下南京市的夫子庙,请用中文回答"
get_completion(query, stream=True)
```
